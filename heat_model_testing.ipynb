```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import detrend
from heat import heat
```

```python
tmin = 1023
tmax = 2023
nt = 199
zmax = 1500
nz = 199
geoflux = 46e-3

t_surf = np.linspace(0,1,nt) + -30

U,t,z = heat(t_surf,
             tmax = tmax,
             tmin = tmin,
             zmax = zmax,
             geoflux = geoflux,
             nz = nz,
             nt = nt,
             accumulation = 0)
```

```python
fig, ax = plt.subplots()
ax.plot(U[:,-1],-z)


fig, ax = plt.subplots()
im = ax.pcolormesh(t,z*-1,U)
fig.colorbar(im, ax=ax)

fig, ax = plt.subplots()
ax.plot(t, U[0,:])
```
