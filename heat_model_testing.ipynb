```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import detrend
from heat import heat
```

```python
tmin = 1923
tmax = 2023
nt = 99
zmax = 200
nz = 199
geoflux = 46e-3

# t_surf = np.linspace(0,1,nt) + -30

t_surf = np.array([-30]*nt) 
```

```python
# With variable conductivity

U_var,t,z = heat(t_surf,
             tmax = tmax,
             tmin = tmin,
             zmax = zmax,
             geoflux = geoflux,
             nz = nz,
             nt = nt,
             accumulation = 0, 
             variable_conductivity=True)

fig, ax = plt.subplots()
ax.plot(U_var[:,-1],z*-1)
ax.set_title('Temperature dependent conductivity')

fig, ax = plt.subplots()
im = ax.pcolormesh(t,z*-1,U_var)
ax.set_title('Temperature dependent conductivity')
fig.colorbar(im, ax=ax)
```

```python
# With constant conductivity

U_con,t,z = heat(t_surf,
             tmax = tmax,
             tmin = tmin,
             zmax = zmax,
             geoflux = geoflux,
             nz = nz,
             nt = nt,
             accumulation = 0, 
             variable_conductivity=False)

fig, ax = plt.subplots()
ax.plot(U_con[:,-1],-z)
ax.set_title('Constant conductivity')

fig, ax = plt.subplots()
im = ax.pcolormesh(t,z*-1,U_con)
fig.colorbar(im, ax=ax)
ax.set_title('Constant conductivity')
```

```python
fig,ax = plt.subplots()

im = ax.pcolormesh(t,z*-1,U_var-U_con)
fig.colorbar(im, ax=ax)
ax.set_title('Difference (variable - constant)')
```
